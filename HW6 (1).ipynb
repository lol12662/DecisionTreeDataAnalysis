```python
Bag_of_words = "bot|follow me|free|fake|virus|troll|clone|tweet me"

dataframe['screen_name_binary']=dataframe['screen_name'].str.contains(Bag_of_words, case = False, na = False)
dataframe['name_binary']=dataframe['name'].str.contains(Bag_of_words, case = False, na = False)
dataframe['description_binary']=dataframe['description'].str.contains(Bag_of_words, case = False, na = False)
dataframe['status_binary']=dataframe['status'].str.contains(Bag_of_words, case = False, na = False)
dataframe.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
df_sub = dataframe#['friends_count','followers_count','screen_name_binary','name_binary','status_binary','verified','statuses_count']
feature_cols =['friends_count','followers_count','statuses_count','screen_name_binary','name_binary','status_binary','verified']
x = df_sub[feature_cols]
y = df_sub['bot']
```

```python
dt = tree.DecisionTreeClassifier(criterion = 'entropy')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
dt = dt.fit(x_train, y_train)
dt.fit(x,y)
y_pred_test = dt.predict(x_test)
y_pred_train = dt.predict(x_train)
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
print("Accuracy test", metrics.accuracy_score(y_test,y_pred_test))
print("Accuracy Train", metrics.accuracy_score(y_train,y_pred_train))
```
